### Handle missing values:
**Cloud9am, Cloud3pm** --> DROP COLUMNS (too many datapoints lost if missing values are dropped (~50%), but probably bad info if nan replaced with average over given period or other methods).

**Evaporation, Sunshine** --> DROP COLUMNS (too many datapoints lost if missing values are dropped (~50%), prefer to drop to avoid having too many datapoints created instead of collected).


**Rainfall, MinTemp, Maxtemp, Temp9am, Humidity9am, Temp3pm, Humidity3pm, Pressure9am, Pressure3pm** --> forward fill  (all these features depend on the period of the year, hence better to use the closets datapoint in time, not average for the entire sample). ***!! this is ok only if we consider the same location, because otherwise we might replace with a value from a different location with very different weather conditions)***


**WindGustSpeed, WindSpeed9am, WindSpeed3pm** --> DROP MISSING VALUES (does average make sense? probably not).

**WindGustDir, WindDir9am, WindDir3pm** --> DROP MISSING VALUES (not numbers, no good way to substitue, we can't know the value, we don't know if nan means no wind or just not recorded data).

**RainToday, RainTomorrow** --> DROP MISSING VALUES (we don't know the value, we can't substitute),

```python
# Perform operations as example, use to visualize data
df = pd.read_csv("data/weatherAUS.csv")

# Create month variable
df["Season"] = (pd.to_datetime(df['Date']).dt.month)

# Map categorical variables to numbers or dummies
dirangle_map = {'N':0, 'NNE':22.5, 'NE':45, 'ENE':67.5, 'E':90, 'ESE': 112.5, 'SE':135, 'SSE':157.5, 'S':180, 'SSW':202.5, 'SW':225, 'WSW':247.5, 'W':270, 'WNW':292.5, 'NW':315, 'NNW':337.5 }
seasons_map = {int(1): "Summer", int(2): "Summer", int(3): "Autumn", int(4):"Autumn", int(5): "Autumn", int(6): "Winter", int(7): "Winter", int(8):"Winter", int(9): "Spring", int(10): "Spring", int(11): "Spring", int(12):"Summer"}
df = df.replace({"WindGustDir": dirangle_map, 
                'WindDir9am':dirangle_map, 
                'WindDir3pm':dirangle_map, 
                'Season': seasons_map,
                'RainToday': {'Yes': 1, 'No': 0}, 
                'RainTomorrow': {'Yes': 1, 'No': 0}})
df = pd.get_dummies(df, columns=["Season"], drop_first=True)

# Drop NaN (here for all columns not dropped)
df = df.dropna(subset=["WindGustSpeed", "WindSpeed9am", "WindSpeed3pm", "WindGustDir", "WindDir9am", "WindDir3pm", "RainToday", "RainTomorrow", "Rainfall", "MinTemp", "MaxTemp", "Temp9am", "Humidity9am", "Temp3pm", "Humidity3pm", "Pressure9am", "Pressure3pm"])
# Drop columns
data = df.drop(["Date", "Location", "Cloud9am", "Cloud3pm", "Evaporation", "Sunshine"], axis=1)
 
# check no missing values
print(f"Any missing values?: {data.isnull().sum().any()}")
print(f"Final dataframe shape: {data.shape}")  
data.head()
```

## Focus on a specific location - Sydeny

```python
# Example on Sydeny
df = pd.read_csv("data/weatherAUS.csv")

df_sydney = df[df["Location"].isin(['Sydney'])]
print(f"Shape: {df_sydney.shape}")
print(f"Missing values:")
print(df_sydney.isnull().sum())
```

```python
# Handel missing values as described earlier (but drop wind variables)
# this time ok because we can fill with values based only on the same location

df_sydney["dtDate"] = pd.to_datetime(df_sydney['Date'])  # for plots 
# Create month variable
df_sydney = df_sydney.sort_values("dtDate")  # sort by dates to be sure
df_sydney["Season"] = (pd.to_datetime(df_sydney['Date']).dt.month)
# Map categorical variables to numbers or dummies
dirangle_map = {'N':0, 'NNE':22.5, 'NE':45, 'ENE':67.5, 'E':90, 'ESE': 112.5, 'SE':135, 'SSE':157.5, 'S':180, 'SSW':202.5, 'SW':225, 'WSW':247.5, 'W':270, 'WNW':292.5, 'NW':315, 'NNW':337.5 }
seasons_map = {int(1): "Summer", int(2): "Summer", int(3): "Autumn", int(4):"Autumn", int(5): "Autumn", int(6): "Winter", int(7): "Winter", int(8):"Winter", int(9): "Spring", int(10): "Spring", int(11): "Spring", int(12):"Summer"}
data_sydney = df_sydney.replace({"WindGustDir": dirangle_map, 
                'WindDir9am':dirangle_map, 'WindDir3pm':dirangle_map, 
                'Season': seasons_map,
                'RainToday': {'Yes': 1, 'No': 0}, 'RainTomorrow': {'Yes': 1, 'No': 0}})
data_sydney = pd.get_dummies(data_sydney, columns=["Season"], drop_first=True)

# Drop NaN (for columns identified above)
data_sydney = data_sydney.dropna(subset=["WindGustSpeed", "WindSpeed9am", "WindSpeed3pm", "WindGustDir", "WindDir9am", "WindDir3pm", "RainToday", "RainTomorrow"])

# Drop columns (for columns identified above)
data_sydney = data_sydney.drop(["Date", "Location", "Cloud9am", "Cloud3pm", "Evaporation", "Sunshine"], axis=1)

# Replace NaN with previous datapoint value (for columns identified above)
data_sydney = data_sydney.fillna(method="ffill", axis=0)
data_sydney = data_sydney.fillna(method="bfill", axis=0)   # required to fill all
 
# check no missing values
print(f"Any missing values?: {data_sydney.isnull().sum().any()}")
print(f"Final dataframe shape: {data_sydney.shape}") 
```

```python
# plot rainfall over time
plt.figure(figsize=(15, 5))
plt.plot(df_sydney.dtDate, df_sydney.Rainfall)
plt.title("Rainfall");
```

```python
# plot other variables over time
fig, axs = plt.subplots(4, 1, figsize=(15, 15))
axs[0].plot(df_sydney.dtDate, df_sydney.Temp3pm)
axs[0].set_title("Temperature 3pm")
axs[1].plot(df_sydney.dtDate, df_sydney.Humidity3pm)
axs[1].set_title("Humidity 3pm")
axs[2].plot(df_sydney.dtDate, df_sydney.Pressure3pm)
axs[2].set_title("Pressure 3pm")
axs[3].plot(df_sydney.dtDate, df_sydney.WindSpeed3pm)
axs[3].set_title("WindSpeed 3pm");

# we can see the seasonality effect, especially for temperature
```

# Create new dataframe with only locations in the South West Area

```python
# Create dataframe
starting_df = pd.read_csv("data/weatherAUS.csv")
target_locs = ["Sydney", "SydneyAirport", "CoffsHarbour", "WaggaWagga", "Wollongong", "MelbourneAirport", "Canberra", "Melbourne", "Richmond", "Bendigo", "Williamtown"]
dataframes_locs = []

# loop over all the locations and create a dataframe for each one (required for correct forward fill)
# then concatenate all and shuffle 
for location in target_locs:
    df = starting_df[starting_df["Location"]==location]
    df = df.sort_values("Date")  
    df["Season"] = (pd.to_datetime(df['Date']).dt.month)
    # Map categorical 
    df = df.replace({'Season': seasons_map,'RainToday': {'Yes': 1, 'No': 0}, 'RainTomorrow': {'Yes': 1, 'No': 0}})
    df = pd.get_dummies(df, columns=["Season"], drop_first=True)
    # Drop NaN (for columns identified above)
    df = df.dropna(subset=["WindGustSpeed", "RainToday", "RainTomorrow"])
    # Drop columns (for columns identified above, also drop all wind variables except WindGustSpeed (distribution ok) in this case)
    df = df.drop(["Date", "Location", "Cloud9am", "Cloud3pm", "Evaporation", "Sunshine", "WindSpeed9am", "WindSpeed3pm", "WindGustDir", "WindDir9am", "WindDir3pm"], axis=1)
    df = df.fillna(method="ffill", axis=0)
    df = df.fillna(method="bfill", axis=0)
    dataframes_locs.append(df)

ausSWdf = pd.concat(dataframes_locs).sample(frac=1)
ausSWdf.shape
```
